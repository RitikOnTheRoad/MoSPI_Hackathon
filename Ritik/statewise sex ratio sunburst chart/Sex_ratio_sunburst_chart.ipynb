```python
#Data used- Estimated number of households and persons by gender and average MPCE for each fractile class of MPCE.
```

```python
import pandas as pd
import plotly.express as px
```

```python
df=pd.read_excel("/content/Estimated number of households and persons by gender and average MPCE for each fractile class of MPCE.xlsx",index_col=[0,1])
```

```python
df
```

```python
df["Total_Males"]=df.Adults_Male+df.Children_Male
df["Total_Females"]=df.Adults_Female+df.Children_Female
```

```python
df.drop(index="All-India",inplace=True)
```

```python
df=df[["Fractile class of MPCE","Total_Males","Total_Females"]]
```

```python
df.reset_index(inplace=True) #multiindex to single index
df
```

```python
df_long=df.melt(id_vars=["State/UT/All-India","Sector","Fractile class of MPCE"],var_name="Gender",value_name="Count")
df_long
```

```python
df_long.Gender=df_long.Gender.map({"Total_Males":"Male","Total_Females":"Female"})
df_long
```

```python
from google.colab import files
```

```python
fig=px.sunburst(df_long,path=["State/UT/All-India","Sector","Fractile class of MPCE","Gender"],values="Count",branchvalues="total")
fig.update_traces(
    insidetextorientation='radial',
    textinfo='label+percent parent',
    marker=dict(line=dict(width=1)) ,

    hovertemplate="<b>%{label}</b><br>"
                 "Total people sampled: %{value}<br>",


     )
fig.update_layout(
    margin=dict(t=10, l=10, r=10, b=10),
    width=1900,
    height=800,
)

fig.write_html("sunburst.html",full_html=True)
files.download("sunburst.html")
```
